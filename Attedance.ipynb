```python
# Function to calculate percentage of class attended
def calculate_attendance(total_days, absent_days):
    attended_days = total_days - absent_days
    attendance_percentage = (attended_days / total_days) * 100
    return attendance_percentage

# Accept input from user
total_days = int(input("Enter the total number of working days: "))
absent_days = int(input("Enter the total number of days absent: "))

# Calculate attendance percentage
attendance_percentage = calculate_attendance(total_days, absent_days)

# Output the result
if attendance_percentage < 75:
    result = f"Attendance Percentage: {attendance_percentage:.2f}%. Student will not be able to sit in the exam."
else:
    result = f"Attendance Percentage: {attendance_percentage:.2f}%. Student is eligible to sit in the exam."

result
```
