```python
from PIL import Image, ImageEnhance, ImageOps, ImageFilter

def apply_vintage_effect(input_image_path, output_image_path):
    try:
        # Open the image
        image = Image.open(input_image_path)
        
        # Convert to grayscale for a base vintage look
        grayscale_image = ImageOps.grayscale(image)
        
        # Apply a sepia tone
        sepia_image = ImageOps.colorize(grayscale_image, "#704214", "#C0A080")
        
        # Add blur for a softer vintage effect
        blurred_image = sepia_image.filter(ImageFilter.GaussianBlur(radius=2))
        
        # Adjust brightness and contrast for a faded look
        enhancer = ImageEnhance.Brightness(blurred_image)
        brightened_image = enhancer.enhance(1.2)
        enhancer = ImageEnhance.Contrast(brightened_image)
        final_image = enhancer.enhance(0.8)
        
        # Save the output
        final_image.save(output_image_path)
        print(f"Vintage effect applied. Saved to: {output_image_path}")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage
input_image = "PKC.jpg"  # Path to the input image
output_image = "output_vintage.jpg"  # Path to save the vintage image

apply_vintage_effect(input_image, output_image)
```

```python
from PIL import Image, ImageEnhance, ImageOps

def apply_vintage_with_cracks(input_image_path, texture_path, output_image_path):
    try:
        # Open the input image
        image = Image.open(input_image_path)
        
        # Open the texture (cracks/noise overlay)
        texture = Image.open(texture_path).convert("RGBA")
        
        # Resize the texture to match the input image
        texture = texture.resize(image.size)
        
        # Convert the input image to sepia
        grayscale = ImageOps.grayscale(image)
        sepia = ImageOps.colorize(grayscale, "#704214", "#C0A080")
        
        # Blend the sepia image with the texture
        sepia = sepia.convert("RGBA")
        vintage_with_texture = Image.blend(sepia, texture, alpha=0.5)  # Adjust alpha for blending strength
        
        # Convert the image to RGB mode to save as JPEG
        final_image = vintage_with_texture.convert("RGB")
        
        # Save the final image
        final_image.save(output_image_path, "JPEG")
        print(f"Vintage image saved as: {output_image_path}")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage
input_image = "PKC.jpg"  # Path to your original image
texture_image = "PKC.jpg"  # Path to the crack/noise texture image
output_image = "output_vintage_cracks.jpg"  # Path to save the final output

apply_vintage_with_cracks(input_image, texture_image, output_image)
```
