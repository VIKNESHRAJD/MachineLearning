## Q20: ATM and Movie Ticket Booking Systems

```python
balance = 1000  

def show_menu():
    print("\nATM Menu:")
    print("1. Check Balance")
    print("2. Deposit")
    print("3. Withdraw")
    print("4. Exit")

def check_balance():
    print(f"Your current balance is: ${balance}")

def deposit():
    global balance
    amount = float(input("Enter amount to deposit: "))
    balance += amount
    print(f"Deposited ${amount}. New balance is ${balance}.")

def withdraw():
    global balance
    amount = float(input("Enter amount to withdraw: "))
    if amount <= balance:
        balance -= amount
        print(f"Withdrew ${amount}. New balance is ${balance}.")
    else:
        print("Insufficient funds.")

while True:
    show_menu()
    choice = input("Enter your choice (1-4): ")
    
    if choice == '1':
        check_balance()
    elif choice == '2':
        deposit()
    elif choice == '3':
        withdraw()
    elif choice == '4':
        print("Exiting ATM. Thank you!")
        break
    else:
        print("Invalid choice. Please try again.")
```

```python
ticket_price = 10.0  

def show_menu():
    print("\nMovie Ticket Booking Menu:")
    print("1. Book Tickets")
    print("2. Check Ticket Price")
    print("3. Exit")

def book_tickets():
    global ticket_price
    number_of_tickets = int(input("Enter the number of tickets to book: "))
    total_cost = number_of_tickets * ticket_price
    print(f"Booked {number_of_tickets} tickets. Total cost is ${total_cost}.")

def check_price():
    print(f"The price per ticket is: ${ticket_price}")

while True:
    show_menu()
    choice = input("Enter your choice (1-3): ")
    
    if choice == '1':
        book_tickets()
    elif choice == '2':
        check_price()
    elif choice == '3':
        print("Exiting booking system. Thank you!")
        break
    else:
        print("Invalid choice. Please try again.")

```

## Q19: Program to Reverse a Number

```python

def reverse_number(num):
    reversed_num = 0
    while num > 0:
        digit = num % 10
        reversed_num = reversed_num * 10 + digit
        num = num // 10
    return reversed_num

# Accept a number from the user
number = int(input("Enter a number to reverse: "))

# Call the function and display the result
reversed_number = reverse_number(number)
print(f"The reversed number is: {reversed_number}")
```
